```python
from nltk.tokenize import word_tokenize
word_tokenize(text)
def tokenize(text):
  return word_tokenize(text)
```

```python
def calculate_positive_score(text):
  temp = tokenize(text)
  score = 0
  for word in temp:
    if word in positive_dictionary:
      score+=1
  return score

def calculate_negative_score(text):
  temp = tokenize(text)
  score = 0
  for word in temp:
    if word in negative_dictionary:
      score+=1
  return score
```

```python
data_links['Positive Score'] = list(map(calculate_positive_score, data_textual))

data_links['Negative Score'] = list(map(calculate_negative_score, data_textual))
```

```python
#Cleaning the Data

def clean(text):
  text = text.replace('?','')
  text = text.replace('!','')
  text = text.replace('.','')
  text = text.replace(',','')
  temp = text.split()
  for word in temp:
    if word in Stop_Words:
      temp.remove(word)
  return temp

def total_words(text):
  temp = clean(text)
  return len(temp)
```

```python
data_links['Total Cleaned Words'] = list(map(total_words, data_textual))
data_links.head()
```

```python
data_links['Subjectivity Score'] = ((temp1 + temp2)/np.array(data_links['Total Cleaned Words'])) + 0.000001
data_links.head()
```

```python
def total_sentences(text):
  sentences = nltk.sent_tokenize(text)
  return len(sentences)
```

```python
data_links['Total Sentences'] = list(map(total_sentences, data_textual))
data_links.head()
```

```python
data_links['Average Sentence Length'] = data_links['Total Cleaned Words']/data_links['Total Sentences']
data_links.head()
```

```python
#Okay I'm just going with the given deinition now
def syllable_count(word):
    count = 0

    vowel = set("aeiouAEIOU")

    #Handling 'es' and 'ed'
    if word[-2:]=='es' or word[-2:]=='ed':
      count = count - 1

    for alphabet in word:
      

        if alphabet in vowel:
            count = count + 1
    
      
    return count

def total_complex_words(text):
  temp = clean(text)
  count = 0
  for word in temp:
    if syllable_count(word)>2:
      count+=1
  return count
```

```python
data_links['Total Complex Words'] = list(map(total_complex_words, data_textual))
data_links.head()
```

```python
data_links['Percentage of Complex words'] = data_links['Total Complex Words']/data_links['Total Cleaned Words']
data_links.head()
```

```python
data_links['Fog Index'] = (data_links['Percentage of Complex words'] + data_links['Average Sentence Length']) * 0.4
data_links.head()
```

```python
data_links['Average Number of Words Per Sentence'] = data_links['Average Sentence Length']
data_links.head()
```

```python
def clean_for_pronouns(text):
  #Also separating the apostrophe
  text = text.replace("'",' ')
  text = text.replace('?','')
  text = text.replace('!','')
  text = text.replace('.','')
  text = text.replace(',','')
  temp = text.split()

  return temp

def total_pronouns(text):
  count = 0

  temp = clean_for_pronouns(text)

  #Not taking US in search
  search = ['I','i','we','We','Ours','ours','My','my','us','Us','Our','our']

  for word in temp:
    if word in search:
      count = count + 1

  return count
```

```python
data_links['Personal Pronouns'] = list(map(total_pronouns, data_textual))
data_links.head()
```

```python
def total_letters(text):
  count = 0
  temp = clean(text)
  for word in temp:
    count = count + len(word)

  return count
```

```python
data_links['Total Letters'] = list(map(total_letters, data_textual))
data_links.head()
```

```python
data_links['Average Word Length'] = data_links['Total Letters']/data_links['Total Cleaned Words']
data_links.head()
```
